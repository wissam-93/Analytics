```python
def filter_satisfaction(x):
  if x <= 0.25:
    return 'very dissatisfied'
  if (x >0.25 and x < 0.5):
    return 'Dissatisfied'
  if (x >= 0.5 and x <=0.75):
    return 'satisfied'
  if (x > 0.75 and x <=100):
    return 'very satisfied'
df['Employee_Satisfaction'] = df['satisfaction'].apply(filter_satisfaction)
```

```python
df.drop(['satisfaction'],axis =1,inplace=True)
```

```python
df.head(5)
```

```python
all_sat = np.round(pd.crosstab(df.Employee_Satisfaction,df.salary),2).reset_index()
all_sat.to_csv('all_sat.csv',index =False)
dept_satisfaction = pd.read_csv('all_sat.csv')
dept_satisfaction
```

```python
dept_satisfaction.groupby(by=['Employee_Satisfaction']).sum()[['high','medium','low']]
```
