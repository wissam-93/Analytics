```python
import pandas as pd
import plotly.express as plt
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('Instagram_after.csv')
```

```python
df.head()
```

```python
fashion = df[df['Category'].str.contains('Fashion',na = False)]
fashion.head()
```

##Most Country audince that follows Fashion Category

```python
fig = plt.bar(fashion, x="Category", y='follow_number', title="Most Country audince that follows Fashion Category",color='Audience_Country',text_auto=True,width = 2000,height=900)
fig.update_traces(textfont_size=20, textangle=5, textposition="outside",cliponaxis=False)
fig.show()
```

##Top 10 Instagramers

```python
top_ten = df.nlargest(10,'follow_number').reset_index()
```

```python
top_ten
```

```python
fig = plt.bar(top_ten, x="Name", y="follow_number", title="Top 10 Instagramers",color='Name',text_auto=True)
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.show()
```

##Top  Athlete or club follower country

```python
sport = df[df['Category'].str.contains('Sports',na = False)]
sport.head()
```

```python
top_sport_followers = sport.nlargest(10,'follow_number').reset_index()
top_sport_followers
```

```python
fig = plt.bar(top_sport_followers, x="follow_number", y='Name',color='Audience_Country',title="Most followed sport industry or athlete by audince country",text_auto=True)
fig.update_traces(textfont_size=20, textangle=5)
fig.show()
```

#Top musicians followers audince country

```python
top_musician = df[df['Category'].str.contains('Music',na = False)]
top_music = top_musician.nlargest(10,'follow_number').reset_index()
```

```python
top_music
```

```python
fig = plt.funnel(top_music, x='follow_number', y='Name',color='Audience_Country',title = 'Top musicians followers audince country')
fig.show()
```
